```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
sns.set_theme()
```

```python
# use sample to generate a random subsample
df = pd.read_csv("../data/Library_Usage.csv",low_memory=False,na_values="Null").sample(n=1000)
```

```python
# Funktion relplot() erzeugt ein Streudiagram zur Visualisierung einer bivariaten Verteilung mit metrischen Variablen
sns.relplot(x='Total Checkouts', y='Total Renewals', data=df)
```

```python
plt.show()
```

```python
# Die Größe des Plots kann über die beiden Argumente height (Höhe in inches) und aspect (Breite des Plots ergibt sich aus aspect * height) 
# konfiguriert werden
sns.relplot(x='Total Checkouts', y='Total Renewals', data=df, height=5, aspect=3)
```

```python
# Um den Plot als Bilddatei abzuspeichern wird die Funktion savefig() aus der matplotlib Bibliothek verwendet
plt.savefig('../tutorial.png', dpi=150)
```

```python
plt.show()
```

```python
# Mit den zusätzlichen Argumenten 
# hue (Farbe der Punkte/ Linien/ Balken …), 
# size (Größe der Punkte/ Linien/ Balken …), 
# style können weitere Eigenschaften des Plots angepasst werden
sns.relplot(x='Total Checkouts', y='Total Renewals', 
                hue='Provided Email Address', style='Within San Francisco County', 
                size='Year Patron Registered', data=df)
```

```python
plt.show()
```

```python
# Plots können mit dem row und/oder col Argument auch anhand einer oder mehrerer Variablen
# in Form von Spalten und/oder Zeilen angeordnet werden
sns.relplot(x='Total Checkouts', y='Total Renewals', 
            size='Year Patron Registered',
            row='Provided Email Address', col='Within San Francisco County', data=df)
```

```python
plt.show()
```

#### Farbpaletten

```python
# Mit dem Argument palette können verschiedene Farbpaletten für den hue Parameter ausgewählt werden
# Zu empfehlen sind die Paletten von ColorBrewer
# (https://s3.amazonaws.com/codecademy-content/programs/dataviz-python/unit-5/seaborn-design-2/article2_image9.png)
sns.relplot(x='Total Checkouts', y='Total Renewals', 
            hue='Provided Email Address', 
            col='Within San Francisco County',
            palette=sns.color_palette('Accent', 2), 
            data=df)
```

```python
plt.show()
```

### Nominale und ordinale Variablen

##### Univariate Häufigkeits- und Bivariate Kreuztatabellen können mit Balkendiagrammen visualisiert werden

```python
sns.catplot(y='Year Patron Registered', data=df, kind='count', color="steelblue")
```

```python
plt.show()
```

```python
sns.catplot(y='Age Range', hue='Provided Email Address', data=df, kind='count')
```

```python
plt.show()
```

```python
sns.catplot(x='Patron Type Definition', 
            data=df, kind='count', 
            col='Year Patron Registered', col_wrap=4)
```

```python
plt.show()
```

### Metrische Variablen

##### Univariate Verteilungen werden mit Histogrammen oder Kernel-Dichte Schätzern visualisiert

```python
# Histogram
sns.displot(df['Total Renewals'], kde=False)
```

```python
plt.show()
```

```python
# With density estimation
sns.displot(df['Total Renewals'], kde=True)
```

```python
plt.show()
```

### Kombination aus metrischen und nominalen/ ordinalen Variablen

```python
# Swarmplot
sns.catplot(x='Year Patron Registered', y = 'Total Renewals',
            data=df, kind='swarm', color="steelblue", aspect=4)
```

```python
plt.show()
```

```python
# Boxplot
sns.catplot(col='Year Patron Registered', y = 'Total Renewals',
            data=df, kind='box', color="steelblue", aspect=4)
```

```python
plt.show()
```
