```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_theme()
```

```python
# Histogramm über die Anzahl der Ausleihen im Datensatz
# Histogram zeigt, dass die Verteilung der Ausleihen durch einige sehr große Ausreißer geprägt ist
sns.displot(df['Total Checkouts'], kind='kde')
```

```python
plt.figure(figsize=(4, 3))
sns.histplot(data=df, x='Total Checkouts', bins=95)
plt.title('Distribution of Total Checkouts')
plt.xlabel('Values of Total Checkouts')
plt.ylabel('')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(4, 2))
sns.kdeplot(data=df, x='Total Checkouts', multiple="stack")
plt.title('Distribution of Total Checkouts')
plt.xlabel('Values of Total Checkouts')
plt.ylabel('')
plt.grid(True)
plt.show()
```

##### Aufgabe 3.7 - Exkurs: Ausreißerentfernung III

Erstelle eine neue Spalte 'Total Checkouts Sqrt', die die Wurzel über die Spalte Total Checkouts enthält. 
Die Wurzel für jede Beobachtung kannst Du mit df['Total Checkouts']**(0.5) berechnen.

```python
df['Total Checkouts Sqrt'] = df['Total Checkouts']**(0.5)
```

```python
df['Total Checkouts Sqrt']
```

Schaue Dir das Histogramm von 'Total Checkouts Sqrt' an und charakterisiere die Verteilung.

```python
plt.figure(figsize=(4, 3))
sns.histplot(data=df, x='Total Checkouts Sqrt', bins=350)
plt.title('Distribution of Total Checkouts Sqrt')
plt.xlabel('Values of Total Checkouts Sqrt')
plt.ylabel('')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(4, 2))
sns.kdeplot(data=df, x='Total Checkouts Sqrt')
plt.title('Distribution of Total Checkouts Sqrt')
plt.xlabel('Values of Total Checkouts Sqrt')
plt.ylabel('')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(4, 2))
sns.displot(data=df, kind="ecdf", x='Total Checkouts Sqrt', rug=True)
plt.title('Distribution of Total Checkouts Sqrt')
plt.xlabel('Values of Total Checkouts Sqrt')
plt.ylabel('')
plt.grid(True)
plt.show()
```
