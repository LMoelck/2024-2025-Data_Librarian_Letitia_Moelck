```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set_theme()
```

```python
df = pd.read_csv("../data/Library_Usage.csv", low_memory=False)
```

```python
df = df[df['Total Checkouts'] < 816]
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(16, 6))
palette = ['plum', 'g', 'orange', 'b', 'r']
plt = sns.boxplot(y='Total Checkouts', x='Year Patron Registered', data=df, fliersize=0.5, whis=1.5, palette=palette)
plt.get_figure().savefig("boxplots.png")
```
