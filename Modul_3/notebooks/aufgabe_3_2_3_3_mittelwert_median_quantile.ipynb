Zur Berechnung des Medians werden zuerst alle Werte in der Spalte "Total Checkouts" in aufsteigender Reihenfolge geordnet.
Danach wird den Wert bestimmt, der die Series genau in zwei Teilen teilt. 
Hat die Series eine gerade Anzahl von Beobachtungen/Ausprägungen, werden dann die zwei Werte, die die Series genau in zwei Teilen teilt, 
addiert und durch 2 geteilt. Das Ergebnis ist der Median. 
Der Median sagt aus, welcher Wert genau in der Mitte einer Datenreihe liegt. 

Zur Berechnung des Mittelwerts werden die einzelen Ausprägungen mit der Anzahl deren Vorkommens multipliziert und durch die Anzahl 
der Beobachtungen geteilt. Das Ergebnis ist der Mittelwert.
    
Für statistische Analyse und Reports ist der Mittelwert aussagekräftig. 
Im Fall der Gesamtausleihen sagt der Mittelwert aus, dass im Durchschnitt 155 Medien ausgeliehen wurden.

The median and IQR are called robust estimates because extreme observations have little effect on their values. 
The mean and standard deviation are much more affected by changes in extreme observations.
The median and IQR are only sensitive to numbers near Q1, the median, and Q3.
Since values in these regions are relatively stable – there aren’t large jumps between observations – the median and IQR estimates are also quite stable.
The interquartile range (IQR, for short) is a measure of variability in data ( like the standard deviation).
The more variable the data, the larger the standard deviation and IQR.
The two boundaries of the box plot are called the first quartile (the 25th percentile, i.e. 25% of the data fall below this value) 
and the third quartile (the 75th percentile), and these are often labeled Q1 and Q3, respectively.

The IQR is the length of the box in a box plot. It is computed as IQR = Q3 − Q1 where Q1 and Q3 are the 25th and 75th percentiles
