```python
# import the package and give it the shorter name 'plt'
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# create some dummy data
x = range(1, 10)
```

```python
# make a simple scatter plot of the data
plt.plot(x, x, c="green", linestyle='', marker='+')
```

```python
# make a simple scatter plot of the data (changed colour)
plt.plot(x, x, c="black", linestyle='', marker='_')
```

```python
plt.plot(x, x, c="black", linestyle='dashed')
```

```python
plt.plot(x, x, c="black")
```
