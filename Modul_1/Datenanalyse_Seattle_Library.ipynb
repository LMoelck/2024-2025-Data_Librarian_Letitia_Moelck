```python
import urllib.request
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Enable inline plotting
%matplotlib inline
```

```python
# Checkout-Daten über die API beziehen und als CSV-Datei speichern
url = "https://data.seattle.gov/resource/tmmm-ytt6.csv?$where=checkoutyear=2018&$limit=10000"
checkout_file = "checkout.csv"
urllib.request.urlretrieve(url, checkout_file)
```

```python
# Checkout-Daten in einen DataFrame laden und analysieren
checkouts_df = pd.read_csv(checkout_file)
```

```python
checkouts_df
```

```python
# Return a tuple representing the dimensionality of the DataFrame (number of columns and rows).
checkouts_df.shape
```

```python
# unique() method is used when we deal with a single column of a DataFrame and returns all unique elements of a column
checkouts_df["checkoutmonth"].unique()
```

```python
# max() method finds the maximum of the values in the object and returns it
checkouts_df["checkouts"].max()
```

```python
# min() method returns the minimum of the values over the requested axis.
checkouts_df["checkouts"].min()
```

```python
# median() method returns a Series with the median value of each column 
# By specifying the column axis (axis='columns'), the median() method searches column-wise and returns the median value for each row
# median = mid point value
checkouts_df["checkouts"].median()
```

```python
# mean() method returns a Series with the mean value of each column.
# By specifying the column axis (axis='columns'), the mean() method searches column-wise and returns the mean value for each row.
# mean = the average value 
checkouts_df["checkouts"].mean()
```

```python
# returns all unique elements of the column "usageclass"
checkouts_df["usageclass"].unique()
```

```python
# returns all unique elements of the column "materialtype"
checkouts_df["materialtype"].unique()
```

```python
# display a part of the dataframe looking for a single value in a certain column
checkouts_df[checkouts_df["materialtype"] == "AUDIOBOOK"]
```
