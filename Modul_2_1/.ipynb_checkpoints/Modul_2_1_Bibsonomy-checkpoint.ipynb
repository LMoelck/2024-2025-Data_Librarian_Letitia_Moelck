```python
# Methode .concat verwenden, um die DataFrames mit den beiden Suchergebnissen miteinander zu kombinieren. 
df_voll = pd.concat([df1_only_publications, df2_only_publications])
```

```python
df_voll
```

```python
# Spalten umbenennen 
df_voll = df_voll.rename(columns={"label": "title", "author": "authors", "booktitle": "published-in"})
```

```python
df_voll
```

```python
df_voll.fillna("k.A") # NaN durch den Wert "k.A." (keine Angabe) ersetzen
```

```python
df_voll = df_voll.fillna("k.A")
```

```python
df_voll
```

```python
df_voll.drop_duplicates(subset = "title", keep = "first") # Duplikate entfernen
```

```python
df_voll["pub-type"].unique()
```

```python
# Filtern nach einem bestimmten Wert in der Spalte "year"
df_voll.loc[df_voll["year"] > 2013]
```

```python
# Mit der Methode .unique() alle verschiedenen Werte in der Spalte "year" ausgeben
df_voll["year"].unique()
```

```python
df_voll_2 = df_voll.apply(lambda x: x.replace({'submitted for publication':'k.A', '2001--':'2001', 'Juli 2019':'2019', ' 2015 ':'2015'}, regex=True))
```

```python
df_voll_2["year"].unique()
```

```python
# Mit der Methode .apply() und "lambda row" wird über jeden Wert in der Spalte "year" zeilenweise iteriert 
# Es wird geprüft, ob die Werte Zahlen sind, dabei werden die Strings in Ganzzahlen umgewandelt
df_voll_2["year_correct"] = df_voll_2.apply(lambda row: row.year.isdigit(), axis=1)
```

```python
df_voll_2
```

```python
# Prüfen, wie viele Werte in der Spalte "year" keine Zahl sind
df_voll_2.loc[df_voll_2.year_correct == False]
```

```python
# Alle Zeilen mit einer Ganzzahl als Wert in der Spalte "year" filtern
df_voll_2 = df_voll_2.loc[df_voll_2.year_correct == True]
```

```python
# Werte in der Spalte "year" mit der Methode .astype() in Integer umwandeln
df_voll_2["year_number"] = df_voll_2["year"].astype(int)
```

```python
df_voll_2.info()
```

```python
# Nicht mehr benötigten Spalten löschen
df_with_year_number = df_voll_2.drop(columns=['year', 'year_correct'])
```

```python
df_with_year_number
```

```python
df_with_year_number["doi"].unique()
```

```python
# Mit Series.replace() die DOI-Angaben vereinheitlichen
df_with_year_number["doi"] = df_with_year_number["doi"].replace('^http.://doi.org/', '', regex=True)
```

```python
df_with_year_number["doi"].unique()
```

```python
# Weitere DOI-Angagen vereinheitlichen
df_with_year_number["doi"] = df_with_year_number["doi"].replace('^http://dx.doi.org/', '', regex=True)
```

```python
df_with_year_number = df_with_year_number.apply(lambda x: x.replace({'http://doi.acm.org/10.1145/1297081.1297091':'10.1145/1297081.1297091'}, regex=True))
```

```python
df_with_year_number["doi"].unique()
```

```python
df_with_year_number["issn"].unique()
```

```python
df_with_year_number = df_with_year_number.apply(lambda x: x.replace({'1548-0666 DOI: 10.4018/IJKM.2019040106':'1548-0666'}, regex=True))
```

```python
df_with_year_number["issn"].unique()
```

```python
df_with_year_number["language"].unique()
```

```python
# Bereinigung der Sprachangaben 
df_with_year_number = df_with_year_number.apply(lambda x: x.replace({'enggg':'en'}, regex=True))
```

```python
df_with_year_number["language"].unique()
```

```python
df_with_year_number
```

```python
df_final = df_with_year_number.rename(columns={"year_number": "year"})
```
