```python
# define how to calculate the idf value for a term
def get_idf_for_term(term, index):
  lex = index.getLexicon()     # get Lexicon
  stemmed_term = stemmer.stem(term)  # get stem of the term

# condition, if term not in Lexicon
  if not stemmed_term in lex:   
    return                    # no return value                
    
  lex_entry = lex[stemmed_term]  # get Lexicon entry

  df_term = lex_entry.getDocumentFrequency()  # for Lexicon entry call the function getDocumentFrequency()
  N = index.getCollectionStatistics().numberOfDocuments   # get number of documents in the collection 
  
  #inverse document frequency
  idf = N/df_term

  #apply logarithm(base 10) to idf
  log_idf = np.emath.logn(10, idf)

  return log_idf
```

```python
# define tf_idf method
# remember that tfidf is the product of two components
# hint: the tf model search result contains tf frequencies

def calc_tf_idf(query, docno, index):
    lex = index.getLexicon()     # get Lexicon
    search_engine = pt.terrier.Retriever(index, wmodel="Tf")
    tfidf_scores = {}

    for term in query.split():  # assumption: query is a String
        stemmed_term = stemmer.stem(term) # get stem of the term
        if stemmed_term in lex:
            lex_entry = lex[stemmed_term]  # get Lexicon entry

            # TF berechnen
            tf_results = search_engine.search(stemmed_term)
            tf = tf_results[tf_results['docno'] == docno]['score'].values[0] if docno in tf_results['docno'].values else 0

            # IDF berechnen
            idf = get_idf_for_term(stemmed_term, index)
            
            # TF-IDF berechnen
            tfidf_scores[term] = tf * idf

    return tfidf_scores
```

```python
query = "python"
docno = "https://www.bibsonomy.org/bibtex/2586c25e1ad4630584da8af63b895871e/fbuckermann"

print(f"The tf-idf for query: {query} and document: {docno} is {calc_tf_idf(query, docno, index)}")
```

```python
#transfrom dataframe into list of dictionaries
ds_dict = df_ds_filtered.to_dict(orient='records')
```

```python
ds_dict
```

```python
# first step: define a new directory for the new index 
index_folder_mult = "./ds_index_mult"

# define the field to be indexed
fields=['docno', 'text', 'tags', 'description']

# define the PyTerrier IterDictIndexer and set the parameters
# Use this Indexer if you wish to index an iter of dicts (possibly with multiple fields).
indexer_mult = pt.IterDictIndexer(index_folder_mult, meta={'docno': 200, 'text': 4096}, fields=True,  text_attrs=fields, overwrite=True)
index_ref_mult = indexer_mult.index(ds_dict) # # Pipeline with the new dictionary with multiple fields to be indexed

print(f"path to our index: {index_ref_mult.toString()}")
```

```python
# load index_ref_mult to an index with the method pt.IndexFactory.of() and get so an python object
index_mult = pt.IndexFactory.of(index_ref_mult)
```

```python
# get again statistics about the dataset
print(index_mult.getCollectionStatistics().toString())
```

```python
search_engine_mult = pt.terrier.Retriever(index_mult, wmodel="TF_IDF")
```
